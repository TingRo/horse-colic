# Download datasets (train, test)

```python
import pandas as pd

columns = ['surgery','Age','Hospital ID','rectal temperature','pulse','respiratory rate','temperature of extremities','peripheral pulse',"mucous membranes",'capillary refill time','pain','peristalsis','abdominal distension','nasogastric tube','nasogastric reflux','nasogastric reflux PH','rectal examination - feces','abdomen',' packed cell volume','total protein','abdominocentesis appearance','abdomcentesis total protein','outcome','surgical lesion?','1','2','3','path']

# train data
horse_file = 'https://raw.githubusercontent.com/TingRo/horse-colic/main/horse-colic.data'
train_data_df = pd.read_csv(horse_file, header=None, na_values='?', sep = ' ', names=columns)
#print(train_data_df.shape)
#print(train_data_df)

# test data
horse_file = 'https://raw.githubusercontent.com/TingRo/horse-colic/main/horse-colic.test'
test_data_df = pd.read_csv(horse_file, header=None, na_values='?', sep = ' ', names=columns)
#print(test_data_df.shape)
#print(test_data_df)
```

```python
# first data                        index
# 2,1,530101,38.50,66,         0
# 28,3,3,?,2,                          5
# 5,4,4,?,?,                           10
# ?,3,5,45.00,8.40,               15
# ?,?,2,2,11300,                   20
# 00000,00000,2                25
```

# Training attribute and label

```python
# train data
horse_train_data = train_data_df.values
#print(horse_train_data.shape)

#remove attribute 3 23 24 25 26 27
idx = [i for i in range(horse_train_data.shape[1]) if i != 2 and i != 22 and i != 23 and i != 24 and i != 25 and i != 26 and i != 27]
# attribute 23
idx_label = [22]
#print(idx)
#print(idx_label)

train_data = horse_train_data[:, idx]
train_label = horse_train_data[:, idx_label]

#print(train_data.tolist())
#print(train_data.shape)
#print(train_label.tolist())
#print(train_label.shape)

# test data
horse_test_data = test_data_df.values
#print(horse_test_data.shape)

test_data = horse_test_data[:, idx]
test_label = horse_test_data[:, idx_label]

#print(test_data.tolist())
#print(test_data.shape)
#print(test_label.tolist())
```

```python
"""
0  surgery
1  age
2  rectal temperature
3  pulse
4  respiratory rate
5  temperature of extremities
6  peripheral pulse
7  mucous membranes
8  capillary refill time
9  pain - a subjective judgement of the horse's pain level
10 peristalsis
11 abdominal distension
12 nasogastric tube
13 nasogastric reflux
14 nasogastric reflux PH
15 rectal examination - feces
16 abdomen
17 packed cell volume
18 total protein
19 abdominocentesis appearance
20 abdomcentesis total protein
"""
```

# KNNImputer

```python
# KNNImputer
# missing value

#print(train_label.shape)

import numpy as np
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors= 7, missing_values=np.NaN, weights='distance')

# train data
imputer_data = imputer.fit_transform(train_data)
imputer_label = imputer.fit_transform(train_label)

#print(imputer_data.tolist())
#print(imputer_data.shape)
#print(imputer_label.shape)

#test data
imputer_test_data = imputer.fit_transform(test_data)
imputer_test_label = imputer.fit_transform(test_label)

#print(imputer_test_data.tolist())
#print(imputer_data.shape)
#print(imputer_test_label.tolist())
#print(imputer_test_label.shape)

# output train imputer result
#imputer_data_DF = pd.DataFrame(imputer_data)
#imputer_data_DF.to_csv('/content/drive/My Drive/Deep Learning/horse/imputer_train.csv',encoding='utf8')
```

# Grouping euthanized with died

```python
#grouping "euthanized" with "died"
#label 1: lived, label 2: died

#when label is nan, round it to int(1 or 2)

# train data
train_label = []
for i in range(imputer_label.shape[0]):
    if imputer_label[i][0]==3:
        imputer_label[i][0] = 2
    elif imputer_label[i][0]==2 or  imputer_label[i][0]==1:
        imputer_label[i][0]=imputer_label[i][0]
    else:
        # round imputer result
        imputer_label[i][0]=round(imputer_label[i][0], 0)
    train_label.append(imputer_label[i][0])

train_label = np.array(train_label)
#print(train_label.tolist())

# test data
test_label = []
for i in range(imputer_test_label.shape[0]):
    if imputer_test_label[i][0]==3:
        imputer_test_label[i][0] = 2
    elif imputer_test_label[i][0]==2 or  imputer_test_label[i][0]==1:
        imputer_test_label[i][0]=imputer_test_label[i][0]
    else:
        # round imputer result
        imputer_test_label[i][0]=round(imputer_test_label[i][0], 0)
    test_label.append(imputer_test_label[i][0])

test_label = np.array(test_label)
#print(test_label.tolist())
```

# KNeighborsClassifier

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors = 15)
model.fit(imputer_data, train_label)
```

# Prediction and accuracy_score

```python
from sklearn.metrics import accuracy_score
test_pred = model.predict(imputer_test_data)
print("accuracy score: ", accuracy_score(test_label, test_pred))
```
